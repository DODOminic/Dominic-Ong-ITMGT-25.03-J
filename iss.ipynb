```python
latitude_list=[]
longitude_list=[]
timestamp_list=[]
import requests
import pprint as pp
import time
for i in range(100):
    page = requests.get("http://api.open-notify.org/iss-now.json")
    page_results=page.json()
    latitude=float(page_results["iss_position"]["latitude"])
    longitude=float(page_results["iss_position"]["longitude"])
    timestamp=page_results["timestamp"]
    latitude_list.append(latitude)
    longitude_list.append(longitude)
    timestamp_list.append(timestamp)
    time.sleep(10)
```

```python
import matplotlib.pyplot as plt 
import numpy as np

plt.plot(latitude_list,longitude_list)
plt.title("ISS Tracker")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(latitude_list, longitude_list, timestamp_list)

plt.title("ISS Tracker")
ax.set_xlabel("Latitude")
ax.set_ylabel("Longitude")
ax.set_zlabel("Timestamp")

plt.show()
```
